```python
import re

fh = open('readme.txt','r')

count = 0

for line in fh.readlines():
    sobj = re.search(r'^\s*$',line)
    # ^ - search at the beginning
    # \s - matches a space
    # * - quantifies previous pattern zero or more times
    # $ - search at the end

    if sobj:
        print(line)
        count += 1

fh.close()
print("number of blank lines {}".format(count))
```
